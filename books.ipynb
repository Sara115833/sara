```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("books.csv")
df
```

```python
df.head()
```

```python
df = df.dropna(subset=['title'])
harry_potter_books = df[df['title'].str.contains('Harry Potter', case=False ,na=False)]
harry_potter_books
```

```python
missing_values = harry_potter_books.isnull().sum()
missing_values
```

```python
harry_potter_df = df[df['title'].str.contains('Harry Potter', case=False)]
most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False)
print("Most Selling Books:")
most_selling_books[['title', 'ratings_count']]
```

```python
average_rating = harry_potter_df['average_rating'].mean()
print("Average:", average_rating)
```

```python
harry_potter_df.drop_duplicates(inplace=True)
harry_potter_df.reset_index(drop=True, inplace=True)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='title', y='books_count', data=harry_potter_df, estimator=sum, ci=None)
plt.xticks(rotation=30, ha='right')
plt.title('Most Selling Books within the Harry Potter Series')
plt.xlabel('Title')
plt.ylabel('Total Books are Sold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=harry_potter_df, x='average_rating', bins=15, kde=True)
plt.title('Distribution of Ratings for Harry Potter Books')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.axvline(average_rating, color='yellow', linestyle='--', label='Average Rating')
plt.legend()
plt.show()
```
